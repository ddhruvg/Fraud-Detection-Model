```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dataset.csv")
```

```python
data.nameOrig.str[0].unique()
```

```python
data.drop(["nameOrig"],inplace=True,axis=1)
```

```python
data.type.unique()
```

```python
data = pd.get_dummies(data,columns=["type"])
```

```python
data.drop(['isFlaggedFraud'],axis=1,inplace=True)
```

```python
data.nameDest.str[0]
```

```python
data.nameDest = data.nameDest.str[0]
```

```python
data = pd.get_dummies(data, ['nameDest'])
```

```python
data.drop(["nameDest_C"],axis=1,inplace=True)
```

```python
data
```

```python
l = ["oldbalanceOrg","newbalanceOrig",'oldbalanceDest','newbalanceDest','amount']

for i in l:
    data[i] = (data[i]-data[i].min())/(data[i].max()-data[i].min())
```

```python
data['step'] = (data['step']-data['step'].min())/(data['step'].max()-data['step'].min())
```

```python
positive_data = data[data.isFraud == 1]
```

```python
negative_data = data[data.isFraud ==0 ]
```

```python
# negative_data = negative_data.sample()
positive_data.shape
```

```python
negative_data = negative_data.sample(8300)
```

```python
negative_data
```

```python
data = pd.concat([positive_data,negative_data])
```

#Neural_Network

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid
%matplotlib widget
import matplotlib.pyplot as plt

import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR)
tf.autograph.set_verbosity(0)
np.set_printoptions(precision=2)
```

```python
model = Sequential(
    [
        Dense(25,activation='relu',name="L1"),
        Dense(10,activation="relu",name='L2'),
        Dense(1,activation='linear',name="L3")
    ],name ="my_model"
)
```

```python
model.summary()
```

```python
from tensorflow.keras.losses import MeanSquaredError
```

```python
model.compile(
    loss=MeanSquaredError(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3)
)
```

```python
y_train
```

```python
X_train=X_train.astype('float64')
```

```python
model.fit(X_train,y_train,epochs=500)
```

```python

logit = model(X_train)
f_x = np.array(tf.nn.sigmoid(logit))
f_x.reshape(1,-1)
for i in range(f_x.shape[0]):
    if f_x[i] >=0.1:
        f_x[i] = 1
    else:
        f_x[i] = 0
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y_train,f_x)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()

Accuracy = metrics.accuracy_score(y_train,f_x)
Precision = metrics.precision_score(y_train,f_x)
F1_score = metrics.f1_score(y_train,f_x)
print(Accuracy,Precision,F1_score)
```

```python
X_test = X_test.astype("float64")
logit = model(X_test)
f_x = np.array(tf.nn.sigmoid(logit))
f_x.reshape(1,-1)
for i in range(f_x.shape[0]):
    if f_x[i] >=0.1:
        f_x[i] = 1
    else:
        f_x[i] = 0

from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y_test,f_x)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()

Accuracy = metrics.accuracy_score(y_test,f_x)
Precision = metrics.precision_score(y_test,f_x)
F1_score = metrics.f1_score(y_test,f_x)
print(Accuracy,Precision,F1_score)        
```

```python
def f1score(y_train,y_predict_train):
    from sklearn import metrics
    confusion_matrix = metrics.confusion_matrix(y_train,y_predict_train)

    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

    cm_display.plot()
    plt.show()

    Accuracy = metrics.accuracy_score(y_train,y_predict_train)
    Precision = metrics.precision_score(y_train,y_predict_train)
    F1_score = metrics.f1_score(y_train,y_predict_train)
    print(Accuracy,Precision,F1_score)
```
